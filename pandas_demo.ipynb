SERIES CREATION
 ‚óè Series can be created using 3 ways : 
    1. using Scalar values
  2. using Dictionary
  3. using Numpy

```python
#1. using Scalar values
series1 = pd.Series(10)
print(series1)
print(type(series1))

#2. using Dictionary
dict1 = {'1':11,'2':12,'3':13}
series2 = pd.Series(dict1)
print(series2)
print(type(series2))

dict1 = {'a':11,'b':12,'c':13}
series2 = pd.Series(dict1)
print(series2)
print(type(series2))

#3. using Numpy
array1 = np.arange(1,10)
number_series = pd.Series(array1)
print(number_series)
print(type(number_series))
```

```python
#access labeled and positional index data
names = ['Akash','Prakash','Riya','Seeya','Shyam','Ram','Meena']
labeled = ['a','b','c','d','e','f','g']
name_series = pd.Series(names,labeled)
print(name_series)
print(type(name_series))
print(name_series['d'])

name_series.index = range(11,18)
print(name_series[13])

new_series = name_series[12:15]
print(new_series)
```

```python
#Rertrieve random elements from the Series
names = ['Akash','Prakash','Riya','Seeya','Shyam','Ram','Meena']

name_series = pd.Series(names)

print(name_series)

print(name_series[2:5])

print(name_series[:5])

print(name_series[3:])

print(name_series[::2])
```

```python
#Series Attributes or properties
#names = ['Akash','Prakash',np.nan,'Seeya','Shyam','Ram','Meena']
names = ['Akash','Prakash','aaa','Seeya','Shyam','Ram','Meena']

name_series = pd.Series(names)

print(name_series)

print(name_series.shape)#(7,)

print(name_series.ndim)#1

print(name_series.size)#7

name_series.name = "NameSeries"#

print(name_series.name)

print(name_series.nbytes)

print(name_series.index)

print(name_series.values)

print(name_series.hasnans)
```

```python
#Series Methods/functions
names = ['Akash','Prakash','aaa','Seeya','Shyam','Ram','Meena']

name_series = pd.Series(names)

print(name_series)

print(name_series.head())#return first 5 series elements
print(name_series.tail())#return last 5 series elements

print(name_series.head(2))#return first n series elements
print(name_series.tail(3))#return last n series elements

print(name_series.count())#return total seroes element
print(name_series.pop(2))#delete given indexed element from the series
print(name_series)

name_series.index = [8,5,1,77,3,4]
print(name_series)
#series elements are sort by index
name_series.sort_index(inplace=True)
print(name_series)
#sorting on series by elements
name_series.sort_values(inplace=True)
print(name_series)
#inplace argument used to add changes in same series
```

```python
series1 = pd.Series([7,1,6,2,4,9,12,6,23,2])
print(series1)
#sort elements in ascending order
asc_series = series1.sort_values()
print(asc_series)
#sort elements in descending order
desc_series = series1.sort_values(ascending=False)
print(desc_series)
```

```python
#Aggregate methods
series1 = pd.Series([7,1,6,2,4,9,12,6,23,2])
print(series1)
print(series1.mean())
print(series1.median())
print(series1.sum())
print(series1.min())
print(series1.max())
print(series1.product())
```

```python
#mathematical operation with scalar value on numerical series
series1 = pd.Series([7,1,6,2,4,9,12,6,23,2])
print(series1)
series1 = series1+10
print(series1)
```
