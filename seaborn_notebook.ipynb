```python
#Load flights dataset to visualize heatmap
flight_df = sns.load_dataset('flights')
flight_df.head()
```

```python
flight_df = flight_df.pivot(index="month", columns="year", 
                            values="passengers")
flight_df.head()
```

```python
#create a simple heatmap
plt.figure(figsize=(12,6))
ax = sns.heatmap(flight_df)
ax
```

```python
#annot : used to display values in cell
plt.figure(figsize=(14,8))
ax = sns.heatmap(flight_df, annot = True, fmt = 'd')
ax
```

```python
#linewidth : is used to add border and border-width for the cells
#linecolor : is used to add different color of border
plt.figure(figsize=(14,8))
ax = sns.heatmap(flight_df, annot = True, fmt = 'd' , linecolor='k', 
                 linewidths='4')
ax
```

```python
#cmap : used different color-combination for the heatmap
plt.figure(figsize=(14,8))
ax = sns.heatmap(flight_df, annot = True, fmt = 'd',
                  linecolor='k', linewidths='5',
                  cmap = 'Reds')
ax
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(flight_df, annot=True, cbar=False, cmap='Blues')
```

```python
#horizantal side color-bar for the heatmap
grid_kws = {"height_ratios": (.3, .05), "hspace": .4}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(flight_df, cbar_kws={"orientation": "horizontal"} , 
                 ax=ax,  cbar_ax=cbar_ax,)
```

```python
#load penguins dataset for pairplot visualization
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
#draw pairplot for penguins dataset
plt.figure(figsize= (12,12))
sns.pairplot(penguins)
```

```python
#pairplot with hue argument to add categories in visualization
plt.figure(figsize= (12,12))
sns.pairplot(penguins, hue = 'sex' , palette='Reds')
```

```python
plt.figure(figsize= (12,12))
sns.pairplot(penguins, hue = 'species' , palette = 'Blues')
```

```python
#kind : is used to dicide type of diagonal charts is it kde plot or histogram
plt.figure(figsize= (12,12))
sns.pairplot(penguins, hue='species',kind = 'hist' )
#values for kind = 'hist' or 'kde'
#add blur effect on scatter plots
```

```python
#diag_kind : used to add histogram for diagonal charts and remaining charts
#displays as scatter charts
plt.figure(figsize = (12,12))
#in pair plot to add histogram used argument diag_kind to hist
sns.pairplot(penguins, hue="species", diag_kind="hist" , 
             palette = 'rainbow')
```

```python
#corner argument : used to add or remove upper diagonal tringle charts from
#pairplot; default value is False
plt.figure(figsize = (12,12))
sns.pairplot(penguins, corner=True)#default corner value is False
```

```python
#markers : set different markers for the categorized data in scatter plot
plt.figure(figsize = (12,12))
sns.pairplot(penguins, hue="sex", markers=["o", "s"] , 
             palette = 'inferno')
```
