```python
#A canvas with multiple Line charts in different colors
import numpy as np
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
x2 = np.arange(1,6)
y2 = np.array([1, 3, 5, 7, 9])
x3 = np.array([1, 2, 3, 4, 5])
y3 = np.array([5, 4, 3, 2, 1])
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.legend(["first","second","third"])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plotting Multiple Lines')
plt.show()
```

```python
#display plot/chart with different canvas size
import numpy as np
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
#plt.figure(figsize=(1.5,1.8))
plt.figure(figsize=(7,5))
plt.plot(x1,y1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot With Figure Size')
plt.show()
```

```python
#plot with linestyle(ls), linecolor(c), linewidth(lw),
#alpha argument
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
x2 = np.arange(1,6)
y2 = np.array([1, 3, 5, 7, 9])
x3 = np.array([1, 2, 3, 4, 5])
y3 = np.array([5, 4, 3, 2, 1])
plt.plot(x1,y1,ls="solid")#linestyle = solid, dashed, dotted
# plt.plot(x1,y1,ls="-")# -, --, -., :
plt.plot(x2,y2,lw="4")#linewidth="3"
#plt.plot(x3,y3,c="red")#linecolor="green"
#linecolor character is r, g, b, k
plt.plot(x3,y3,"r",alpha=0.4)#alpha used to add brightness or
#transperancy to the plot
plt.legend(["first","second","third"])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plotting Multiple Lines')
plt.show()
```

```python
#line plot with grids 
#applye different style on grids
x1 = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 6, 8, 10])
plt.plot(x1,y1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot With Figure Size')
#plt.grid(axis="both")#x,y,both
plt.grid(axis="both",ls="--",lw=1,c="red")
plt.show()
```

```python
#subplot() : used to create both a figure and axes 
x = [1,2,3,4,5]
y = [50,40,30,20,10]
plt.subplot(2,1,1)#(row,column,index/position)
plt.plot(x,y,'b')
plt.subplot(2,1,2)
plt.plot(y,x,'g')
plt.show()
```

```python
#subplot() : used to create both a figure and axes 
x = [1,2,3,4,5]
y = [50,40,30,20,10]
plt.subplot(1,2,2)#(row,column,index/position)
plt.plot(x,y,'b')
plt.subplot(1,2,1)
plt.plot(y,x,'g')
plt.show()
```

```python
#subplots()
#add grid to the canvas
x = np.array([10,20,30,40])

fig, axes = plt.subplots(1, 2, figsize=(10,5))
#it return canvas instance and array of subplots

axes[0].plot(x,x+1,color="pink",lw=1,marker="1")
axes[0].grid(True)
axes[0].set_title("Grid Line Plot")

axes[1].plot(x, x+5, color='blue', marker='o', markersize=10)
```

```python
#add grid to the canvas and add multiple rows and columns
# with title for each plot
x = np.array([10,20,30,40])

fig, axes = plt.subplots(2, 2, figsize=(10,5))

axes[0,0].plot(x,x+1,color="pink",lw=1,marker="1")
axes[0,0].set_title("Plot1")

axes[1,0].plot(x, x+5, color='blue', marker='o', markersize=10)
axes[1,0].set_title("Plot2")

axes[0,1].plot(x,x+1,color="pink",lw=1,marker="1")
axes[0,1].set_title("Plot3")

axes[1,1].plot(x, x+5, color='blue', marker='o', markersize=10)
axes[1,1].set_title("Plot4")
```
